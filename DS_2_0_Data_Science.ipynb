## 1.0 Code Task - Average of Rows
In a matrix, or 2-d array X, the averages (or means) of the elements of rows is called row means.

**Task**

Given a 2D array, return the rowmeans.

**Input Format**

First line: two integers separated by spaces, the first indicates the rows of matrix X (n) and the second indicates the columns of X (p)

Next n lines: values of the row in X

**Output Format**

An numpy 1d array of values rounded to the second decimal.

```python
n, p = [int(x) for x in input().split()]

import numpy as np 
# create an empty list
row= []
for i in range(n):
# appends the inputed value to the empty list
    row.append(input().split())

# converts the list to a numpy array
# converts the array values to a float16
#  finds the mean and rounds it off
print(np.array(row).astype(np.float16).mean(axis=1).round(2))
```

# - Linear Regression (LR)

LR is a simple supervised learning model. It is used when there is a linear relationship between features in a dataset. It fits a straight line to data. In elementary maths, equation of a straight line is given by:

**y = mx + b**

b - intercept

m - slope

x - a feature ( given a dataset, a crucial column)

y - label/ output ( what is being predicted)

The goal of LR is to find m, and b so as to minimise errors. 

Given these points, (1.5, 2.8), (2, 5.3), (2.5, 5.5), (3, 7), (3.5, 8.8): A straight line would be realised as:

image.png


Then we try to draw a line through the points from several angles:






image.png




The red seems to be the closest line of best fit. We now consider the outliers around the red line, in an attempt to capture the distance between them and the line. The difference is called ** Residuals**.

 Linear regression essentially finds the line that minimizes the sum of the squared residuals.

## Fiting a Univariate Linear Regression
1. Instantiate a model

2. Train-test split- we will split the data into a training set and testing set. The whole idea is to teach the machine using the training data, then we test the machine with the test data. We split the entire dataset into a 70-30 ratio with the highest going into training. We set the random state so that the results will be reproducible.

3. Fit the model - fitting the model is equal to training it. it fits the model to the data and finds the co-efficients we are looking for. After here, the intercept and the slope are ready

4. Prediction - using the predict method ( .predict()) we use an average number of room 6.5

5. Evaluating the model - examine the model performance by comparing the fitted line and the true observations. From the graph we look out for the outliers and measure the residuals. **A residual is the difference between the observed value of the target and the predicted value.** The closer the residual is to the 0, the better the model. There are several ways of evaluating a model, they include:

- **Root mean squared error (RMSE)** - tells you how concentrated the data is around the line of best fit.

- **R-squared** - its proportion explains the variability of model. The total variation is calculated as the sum of squares of the difference between the response and the mean of response, in the example of testing data:


Between RMSE and R-squared the RMSE tells the concentration of the data in absolute terms, while the R-squared value does it in percentage terms

```python
# import the Linear regression class and instantiate the model
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
# we split the data into training 
# and testing sets

from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, 
  test_size = 0.3, 
  random_state=1)

# print the shapes of the datasets

print(X_train.shape)
print(Y_train.shape)
print(X_test.shape)
print(Y_test.shape)
```

```python
# fitting the data
# training the data
model.fit(X_train, Y_train)
```

```python
# intercept
model.intercept_.round(2)
```

```python
# slope
model.coef_.round(2)
```

 with intercept and slope being -30.57 and 8.46, our equation now becomes
 MEDV = -30.57 + 4.46*RM

```python
# prediction

new_RM = np.array([6.5]).reshape(-1,1)
model.predict(new_RM)
```

Note that the input has to be 2-dimensional, either a 2darray or DataFrame will work in this case.

```python
# feed the testing test
# and get predictions on all rooms
y_test_predicted = model.predict(X_test)
print(y_test_predicted.shape)
print(type(y_test_predicted))
```

```python
# evaluating the model
plt.scatter(X_test, Y_test, label='Testing data')
plt.plot(X_test, y_test_predicted,
label='prediction', linewidth=3)
plt.xlabel('RM'); plt.ylabel('MEDV')
plt.legend(loc='upper left')
plt.savefig("plot.png")
plt.show()
```

```python
# calculating and plotting the residual
residuals = Y_test - y_test_predicted

# plot residuals
plt.scatter(X_test, residuals)
# plot a horizontal line at y = 0
plt.hlines(y = 0,
xmin = X_test.min(), xmax=X_test.max(),
linestyle='--')
# set xlim
plt.xlim((4, 9))
plt.xlabel('RM'); plt.ylabel('residuals')
plt.savefig("plot.png")
plt.show()
```

How to know that your model is optimum:

1. Ideally the residuals should be symmetrically and randomly spaced around the horizontal axis
* if there's a pattern, linear or non-linear, it is an indication that the model could do with some improvements


There is a need to find a single term that would represent the residual of the model. We could use the mean, but since the mean has both negative values and positive values, they will cancel out. and so we explore the mean_squared error

```python
residuals[:5]
```

```python
# using the mean_squared error
from sklearn.metrics import mean_squared_error
mean_squared_error(Y_test, y_test_predicted)
```

the smaller the mean_squared_error, the better but it is for us to interpret. We define it based on the dependent variable MEDV in the test set. Y_test ranges from 6.3 to 50 with a variance of 92.26. And so this is an okay model.

```python
# we calculate the r-squared via the model score
model.score(X_test, Y_test)
```
